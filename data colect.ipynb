```python
import os
import uuid
import time
import matplotlib.pyplot as plt
import cv2
import torch
import numpy as np
```

```python
model=torch.hub.load('ultralytics/yolov5','yolov5s')
```

```python
labels=['Hi','win','love']
num_of_pic=5
```

```python
IMAGE_PATH=os.path.join('Data','images')
IMAGE_PATH
```

```python
for label in labels:
    path=os.path.join(IMAGE_PATH,label)
    if not os.path.exists(path):
        !mkdir {path}
```

```python
cap=cv2.VideoCapture(0)
for label in labels:
    print(f'Capturing image for{label}')
    time.sleep(5)
    for img_num in range(num_of_pic):
        print(f'image no{img_num}')
        rat,frame=cap.read()
        image_name=os.path.join(IMAGE_PATH,label,label+'.'+f'{str(uuid.uuid1())}.jpg')
        cv2.imwrite(image_name,frame)
        cv2.imshow('Sign',frame)
        time.sleep(2)
cap.release()
cv2.destroyAllWindows()
```
